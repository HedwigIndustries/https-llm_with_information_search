# Задание 4.1

Реализуйте функцию `search_course`, которая на вход будет принимать строку `prompt` и число `n`. Где `prompt` - это строка (запрос пользователя), а `n` - это целое неотрицательное число, означающее количество дисциплин, которые нужно вывести по этому запросу. Выведенные дисциплины должны быть отранжированы по релевантности запросу.

```python
!pip install pymorphy2
import pymorphy2
```

```python
import ast
df_3_lab = df_3_lab.head(5000)
df_3_lab['norm'] = df_3_lab['norm'].apply(ast.literal_eval)
```

```python
one_row_per_course = df_3_lab.groupby('course_id').first().reset_index()
```

```python
morph = pymorphy2.MorphAnalyzer()

def preprocess_text(text):
    text = text.lower()
    tokens = text.split()
    lemmas = [morph.parse(token)[0].normal_form for token in tokens if token.isalnum() and len(token)>=4]
    return lemmas

def search_course(prompt, n, df, column_name):
    prompt_processed = preprocess_text(prompt)
    df['relevance'] = df[column_name].apply(lambda text : sum(token in prompt_processed for token in text))
    df_sorted = df.sort_values(by='relevance', ascending=False)
    return df_sorted.head(n)
```

```python
prompt = "архитектура информационных систем"
n = 5
result = search_course(prompt, n, df_3_lab, 'norm')
result.head()
```

```python
prompt = "архитектура информационных систем"
n = 5
result = search_course(prompt, n, one_row_per_course, 'norm')
result.head()
```

```python
prompt = "мышление"
n = 10
result = search_course(prompt, n, one_row_per_course, 'norm')
result.head()
```

# Задание 4.2

Протестируйте функцию `search_course` на датасете из задания 3.2, используя булев поиск по столбцу `norm`, предварительно обработав `prompt`

```python
def search_course_bool(prompt, n, df):
    prompt_processed = preprocess_text(prompt)
    condition = df['norm'].apply(lambda lemmas: all(word in lemmas for word in prompt_processed))
    filtered_df = df[condition]
    filtered_df['relevance'] = filtered_df['norm'].apply(lambda text : sum(token in prompt_processed for token in text))
    df_sorted = filtered_df.sort_values(by='relevance', ascending=False)
    return df_sorted.head(n)
```

```python
prompt = "архитектура информационных систем"
n = 5
result = search_course_bool(prompt, n, df_3_lab)
result.head()
```

# Задание 4.3

Протестируйте функцию `search_course` на датасете, который был загружен выше, используя **tf-idf**.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

morph = pymorphy2.MorphAnalyzer()

def search_course_tfidf(prompt, n, df):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

    prompt_vector = tfidf_vectorizer.transform(preprocess_text(prompt))

    similarity_scores = cosine_similarity(prompt_vector, tfidf_matrix)

    sorted_indices = similarity_scores.argsort()[0][::-1]
    top_n_courses = df.iloc[sorted_indices[:n]]
    return top_n_courses
```

```python
prompt = "архитектура информационных систем"
n = 5
results = search_course_tfidf(prompt, n, df_3_lab)
results.head()
```
